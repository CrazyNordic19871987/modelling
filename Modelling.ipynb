Альтернативная гипотеза: Средние баллы студентов, обучавшихся по методике X, и студентов, обучавшихся по методике Y, статистически значимо различаются.

```python
# Ваш код здесь

import pandas as pd
from scipy import stats
import numpy as np

print("=== АНАЛИЗ ДАННЫХ ДЛЯ ЗАДАЧИ 1 ===")
print("Сравнение методик обучения X и Y\n")

# Создаем данные напрямую из предоставленной таблицы
data = {
    'X': [57, 67, 58, 55, 54],
    'Y': [64, 67, 62, 59, 65]
}

df = pd.DataFrame(data)

print("Данные:")
print(df)
print()

# Извлекаем данные по группам
group_x = df['X']
group_y = df['Y']

print("=== ОПИСАТЕЛЬНАЯ СТАТИСТИКА ===")
print(f"Группа X (методика X): {group_x.tolist()}")
print(f"Группа Y (методика Y): {group_y.tolist()}")
print()

# Основные статистики
mean_x = np.mean(group_x)
mean_y = np.mean(group_y)
std_x = np.std(group_x, ddof=1)
std_y = np.std(group_y, ddof=1)

print(f"Группа X: среднее = {mean_x:.2f}, стандартное отклонение = {std_x:.2f}")
print(f"Группа Y: среднее = {mean_y:.2f}, стандартное отклонение = {std_y:.2f}")
print(f"Разница средних (X - Y): {mean_x - mean_y:.2f} баллов")
print()

# Проверяем равенство дисперсий с помощью теста Левене
print("=== ПРОВЕРКА ПРЕДПОЛОЖЕНИЙ ===")
levene_test = stats.levene(group_x, group_y)
print(f"Тест Левене на равенство дисперсий:")
print(f"p-value = {levene_test.pvalue:.3f}")

# Выбираем тип t-теста
if levene_test.pvalue > 0.05:
    print("Дисперсии можно считать равными -> используем t-тест Стьюдента")
    t_test = stats.ttest_ind(group_x, group_y, equal_var=True)
else:
    print("Дисперсии нельзя считать равными -> используем t-тест Уэлча")
    t_test = stats.ttest_ind(group_x, group_y, equal_var=False)

print()
print("=== РЕЗУЛЬТАТЫ T-ТЕСТА ===")
print(f"t-статистика = {t_test.statistic:.3f}")
print(f"p-value = {t_test.pvalue:.3f}")

print()
print("=== ИНТЕРПРЕТАЦИЯ РЕЗУЛЬТАТОВ ===")
print("Уровень значимости: α = 0.05")
print("Нулевая гипотеза H₀: Средние баллы групп X и Y не различаются")
print("Альтернативная гипотеза H₁: Средние баллы групп X и Y различаются")
print()

if t_test.pvalue < 0.05:
    print(f"p-value ({t_test.pvalue:.3f}) < 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМЫЕ РАЗЛИЧИЯ ОБНАРУЖЕНЫ")
    print("✓ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print("✓ ПРИНИМАЕМ альтернативную гипотезу H₁")
    print()
    print("ВЫВОД: Существуют статистически значимые различия в оценках")
    print("между группой, которая училась по методике X, и группой,")
    print("которая училась по методике Y.")
else:
    print(f"p-value ({t_test.pvalue:.3f}) ≥ 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМЫХ РАЗЛИЧИЙ НЕ ОБНАРУЖЕНО")
    print("✓ НЕ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print()
    print("ВЫВОД: Нет статистически значимых различий в оценках")
    print("между группой, которая училась по методике X, и группой,") 
    print("которая училась по методике Y.")

print()
print("=== ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ ===")
print(f"Средний балл группы X: {mean_x:.2f}")
print(f"Средний балл группы Y: {mean_y:.2f}")
print(f"Разница средних: {mean_x - mean_y:.2f} баллов")

if mean_x > mean_y:
    print("Методика X показала лучшие результаты")
elif mean_x < mean_y:
    print("Методика Y показала лучшие результаты")
else:
    print("Результаты одинаковые")

print("\n" + "="*50)
print("=== АНАЛИЗ ДАННЫХ ДЛЯ ЗАДАЧИ 2 ===")
print("Эффективность добавки для снижения веса\n")

# Создаем примерные данные для задачи 2
np.random.seed(42)
n_participants = 60

# Вес до приема добавки
weight_before = np.random.normal(loc=85, scale=10, size=n_participants)
# Вес после приема добавки (предполагаем снижение)
weight_after = weight_before - np.random.normal(loc=2.5, scale=1.2, size=n_participants)

# Создаем DataFrame для задачи 2
df_weight = pd.DataFrame({
    'weight_before': weight_before,
    'weight_after': weight_after
})
df_weight['difference'] = df_weight['weight_before'] - df_weight['weight_after']

print("Первые 10 строк данных:")
print(df_weight.head(10))
print()

print("=== ОПИСАТЕЛЬНАЯ СТАТИСТИКА ===")
print(f"Вес ДО приема добавки: {df_weight['weight_before'].mean():.2f} ± {df_weight['weight_before'].std():.2f} кг")
print(f"Вес ПОСЛЕ приема добавки: {df_weight['weight_after'].mean():.2f} ± {df_weight['weight_after'].std():.2f} кг")
print(f"Среднее снижение веса: {df_weight['difference'].mean():.2f} ± {df_weight['difference'].std():.2f} кг")
print(f"Количество участников: {n_participants}")

print(f"\n=== ФОРМУЛИРОВКА ГИПОТЕЗ ===")
print("Нулевая гипотеза H₀: Добавка не влияет на вес, среднее изменение веса = 0")
print("Альтернативная гипотеза H₁: Добавка уменьшает вес, среднее изменение веса > 0")

print(f"\n=== ПРОВЕРКА ГИПОТЕЗ ===")
# Парный t-тест для зависимых выборок
t_test_paired = stats.ttest_rel(df_weight['weight_before'], df_weight['weight_after'], alternative='greater')

print(f"Результаты парного t-теста:")
print(f"t-статистика = {t_test_paired.statistic:.4f}")
print(f"p-value = {t_test_paired.pvalue:.4f}")
print(f"Степени свободы = {n_participants - 1}")

print(f"\n=== ИНТЕРПРЕТАЦИЯ РЕЗУЛЬТАТОВ ===")
print("Уровень значимости: α = 0.05")

if t_test_paired.pvalue < 0.05:
    print(f"p-value ({t_test_paired.pvalue:.4f}) < 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМОЕ УМЕНЬШЕНИЕ ВЕСА ОБНАРУЖЕНО")
    print("✓ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print("✓ ПРИНИМАЕМ альтернативную гипотезу H₁")
    print(f"\nВЫВОД: Добавка статистически значимо уменьшает вес.")
    print(f"Среднее снижение веса составило {df_weight['difference'].mean():.2f} кг")
else:
    print(f"p-value ({t_test_paired.pvalue:.4f}) ≥ 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМОГО УМЕНЬШЕНИЯ ВЕСА НЕ ОБНАРУЖЕНО")
    print("✓ НЕ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print(f"\nВЫВОД: Нет статистически значимых доказательств того,")
    print(f"что добавка уменьшает вес.")

# Дополнительная информация
decreased_count = (df_weight['difference'] > 0).sum()
percentage = (decreased_count / n_participants) * 100
print(f"\nУ {decreased_count} из {n_participants} участников ({percentage:.1f}%) вес уменьшился")
```

Нулевая гипотеза: Добавка не влияет на вес, среднее изменение веса равно 0.

Альтернативная гипотеза: Добавка уменьшает вес, среднее изменение веса больше 0.

```python
# Ваш код здесь

# Установка scipy
!pip install scipy

import pandas as pd
import numpy as np
from scipy import stats

print("=== АНАЛИЗ ЭФФЕКТИВНОСТИ ДОБАВКИ ДЛЯ СНИЖЕНИЯ ВЕСА ===\n")

# Создаем данные для анализа (примерные данные)
np.random.seed(42)

# Вес до приема добавки (в кг)
weight_before = np.array([85.5, 92.3, 78.9, 88.1, 76.4, 94.2, 81.7, 89.5, 83.2, 90.1,
                         87.3, 79.8, 91.6, 84.7, 82.9, 88.9, 80.3, 86.7, 83.5, 89.8])

# Вес после приема добавки
weight_after = np.array([83.2, 90.1, 76.8, 85.9, 74.5, 91.8, 79.4, 87.2, 81.1, 87.9,
                        85.0, 77.6, 89.3, 82.4, 80.7, 86.5, 78.2, 84.3, 81.2, 87.5])

# Создаем DataFrame
df = pd.DataFrame({
    'before': weight_before,
    'after': weight_after
})

# Рассчитываем разницу
df['difference'] = df['before'] - df['after']

print("Данные:")
print(df.head(10))
print()

print("=== ОПИСАТЕЛЬНАЯ СТАТИСТИКА ===")
print(f"Вес ДО приема добавки:")
print(f"  Среднее: {df['before'].mean():.2f} кг")
print(f"  Стандартное отклонение: {df['before'].std():.2f} кг")

print(f"\nВес ПОСЛЕ приема добавки:")
print(f"  Среднее: {df['after'].mean():.2f} кг")
print(f"  Стандартное отклонение: {df['after'].std():.2f} кг")

print(f"\nИзменение веса:")
print(f"  Среднее снижение: {df['difference'].mean():.2f} кг")
print(f"  Стандартное отклонение: {df['difference'].std():.2f} кг")
print(f"  Минимальное снижение: {df['difference'].min():.2f} кг")
print(f"  Максимальное снижение: {df['difference'].max():.2f} кг")
print(f"  Количество участников: {len(df)} человек")

print(f"\n=== ФОРМУЛИРОВКА ГИПОТЕЗ ===")
print("Нулевая гипотеза (H₀): Добавка не влияет на вес, среднее изменение веса = 0")
print("Альтернативная гипотеза (H₁): Добавка уменьшает вес, среднее изменение веса > 0")

print(f"\n=== ПРОВЕРКА ГИПОТЕЗ ===")
print("Используем парный t-тест (paired t-test)")

# Проводим парный t-тест
# alternative='greater' проверяет, что вес уменьшился
t_stat, p_value = stats.ttest_rel(df['before'], df['after'], alternative='greater')

print(f"\nРезультаты парного t-теста:")
print(f"t-статистика = {t_stat:.4f}")
print(f"p-value = {p_value:.4f}")
print(f"Степени свободы = {len(df) - 1}")

print(f"\n=== ИНТЕРПРЕТАЦИЯ РЕЗУЛЬТАТОВ ===")
print(f"Уровень значимости: α = 0.05")

if p_value < 0.05:
    print(f"p-value ({p_value:.4f}) < 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМОЕ УМЕНЬШЕНИЕ ВЕСА ОБНАРУЖЕНО")
    print("✓ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print("✓ ПРИНИМАЕМ альтернативную гипотезу H₁")
    print(f"\nВЫВОД: Добавка статистически значимо уменьшает вес.")
    print(f"Среднее снижение веса составило {df['difference'].mean():.2f} кг")
else:
    print(f"p-value ({p_value:.4f}) ≥ 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМОГО УМЕНЬШЕНИЯ ВЕСА НЕ ОБНАРУЖЕНО")
    print("✓ НЕ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print(f"\nВЫВОД: Нет статистически значимых доказательств того,")
    print(f"что добавка уменьшает вес.")

# Дополнительная информация
print(f"\n=== ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ ===")
# Доверительный интервал
confidence = 0.95
n = len(df)
mean_diff = df['difference'].mean()
std_diff = df['difference'].std()
se = std_diff / np.sqrt(n)
t_critical = stats.t.ppf((1 + confidence) / 2, n - 1)
ci_lower = mean_diff - t_critical * se
ci_upper = mean_diff + t_critical * se

print(f"95% доверительный интервал для среднего изменения веса:")
print(f"[{ci_lower:.2f}, {ci_upper:.2f}] кг")

# Процент участников с уменьшением веса
decreased_count = (df['difference'] > 0).sum()
percentage = (decreased_count / len(df)) * 100
print(f"У {decreased_count} из {len(df)} участников ({percentage:.1f}%) вес уменьшился")
```

```python
# Ваш код здесь

import pandas as pd
import numpy as np
from scipy import stats

print("=== АНАЛИЗ ЭФФЕКТИВНОСТИ УВЕДОМЛЕНИЙ ДЛЯ ПОВЫШЕНИЯ АКТИВНОСТИ ===\n")

# Создаем примерные данные активности пользователей
np.random.seed(42)

# Активность до уведомлений (посещений в неделю)
activity_before = np.random.poisson(lam=2.5, size=50)  # в среднем 2.5 посещения в неделю

# Активность после уведомлений (предполагаем увеличение)
activity_after = activity_before + np.random.poisson(lam=1.2, size=50)

# Создаем DataFrame
df = pd.DataFrame({
    'before': activity_before,
    'after': activity_after
})

# Рассчитываем разницу
df['difference'] = df['after'] - df['before']

print("Первые 10 строк данных:")
print(df.head(10))
print()

print("=== ОПИСАТЕЛЬНАЯ СТАТИСТИКА ===")
print(f"Активность ДО уведомлений:")
print(f"  Среднее: {df['before'].mean():.2f} посещений/неделю")
print(f"  Стандартное отклонение: {df['before'].std():.2f}")
print(f"  Медиана: {df['before'].median():.2f}")
print(f"  Минимум: {df['before'].min()}")
print(f"  Максимум: {df['before'].max()}")

print(f"\nАктивность ПОСЛЕ уведомлений:")
print(f"  Среднее: {df['after'].mean():.2f} посещений/неделю")
print(f"  Стандартное отклонение: {df['after'].std():.2f}")
print(f"  Медиана: {df['after'].median():.2f}")
print(f"  Минимум: {df['after'].min()}")
print(f"  Максимум: {df['after'].max()}")

print(f"\nИзменение активности:")
print(f"  Среднее увеличение: {df['difference'].mean():.2f} посещений/неделю")
print(f"  Стандартное отклонение: {df['difference'].std():.2f}")
print(f"  Минимальное изменение: {df['difference'].min()}")
print(f"  Максимальное изменение: {df['difference'].max()}")
print(f"  Количество пользователей: {len(df)}")

print(f"\n=== ФОРМУЛИРОВКА ГИПОТЕЗ ===")
print("Нулевая гипотеза (H₀): Уведомления не влияют на активность пользователей,")
print("                       среднее изменение активности = 0")
print("Альтернативная гипотеза (H₁): Уведомления увеличивают активность пользователей,")
print("                             среднее изменение активности > 0")

print(f"\n=== ПРОВЕРКА ГИПОТЕЗ ===")

# Проверяем нормальность распределения разниц
shapiro_test = stats.shapiro(df['difference'])
print(f"Проверка нормальности распределения разниц (тест Шапиро-Уилка):")
print(f"p-value = {shapiro_test.pvalue:.3f}")

if shapiro_test.pvalue > 0.05:
    print("Распределение разниц можно считать нормальным -> используем парный t-тест")
    # Парный t-тест для зависимых выборок
    t_test = stats.ttest_rel(df['after'], df['before'], alternative='greater')
    test_used = "парный t-тест"
else:
    print("Распределение разниц не нормальное -> используем критерий Уилкоксона")
    # Критерий Уилкоксона для парных выборок
    t_test = stats.wilcoxon(df['after'], df['before'], alternative='greater')
    test_used = "критерий Уилкоксона"

print(f"\nРезультаты статистического теста ({test_used}):")

if test_used == "парный t-тест":
    print(f"t-статистика = {t_test.statistic:.4f}")
    print(f"p-value = {t_test.pvalue:.4f}")
    print(f"Степени свободы = {len(df) - 1}")
else:
    print(f"Статистика = {t_test.statistic:.4f}")
    print(f"p-value = {t_test.pvalue:.4f}")

print(f"\n=== ИНТЕРПРЕТАЦИЯ РЕЗУЛЬТАТОВ ===")
print("Уровень значимости: α = 0.05")

if t_test.pvalue < 0.05:
    print(f"p-value ({t_test.pvalue:.4f}) < 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМОЕ УВЕЛИЧЕНИЕ АКТИВНОСТИ ОБНАРУЖЕНО")
    print("✓ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print("✓ ПРИНИМАЕМ альтернативную гипотезу H₁")
    print(f"\nВЫВОД: Ежедневные уведомления статистически значимо увеличивают")
    print(f"активность пользователей.")
    print(f"Среднее увеличение активности составило {df['difference'].mean():.2f} посещений в неделю")
else:
    print(f"p-value ({t_test.pvalue:.4f}) ≥ 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМОГО УВЕЛИЧЕНИЯ АКТИВНОСТИ НЕ ОБНАРУЖЕНО")
    print("✓ НЕ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print(f"\nВЫВОД: Нет статистически значимых доказательств того,")
    print(f"что уведомления увеличивают активность пользователей.")

print(f"\n=== ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ ===")

# Процент пользователей с увеличением активности
increased_count = (df['difference'] > 0).sum()
no_change_count = (df['difference'] == 0).sum()
decreased_count = (df['difference'] < 0).sum()

print(f"Пользователей с увеличением активности: {increased_count} ({increased_count/len(df)*100:.1f}%)")
print(f"Пользователей без изменений: {no_change_count} ({no_change_count/len(df)*100:.1f}%)")
print(f"Пользователей с уменьшением активности: {decreased_count} ({decreased_count/len(df)*100:.1f}%)")

# Эффективность в абсолютных числах
print(f"\nАбсолютные изменения:")
print(f"Общее увеличение посещений в неделю: {df['difference'].sum()}")
print(f"Среднее увеличение на пользователя: {df['difference'].mean():.2f} посещений/неделю")

# Расчет относительного изменения
relative_increase = ((df['after'].mean() - df['before'].mean()) / df['before'].mean()) * 100
print(f"Относительное увеличение активности: {relative_increase:.1f}%")

# Анализ пользователей, которые стали посещать ежедневно (7+ раз в неделю)
daily_users_before = (df['before'] >= 7).sum()
daily_users_after = (df['after'] >= 7).sum()
print(f"\nПользователей с ежедневной активностью (7+ посещений/неделю):")
print(f"До уведомлений: {daily_users_before} ({daily_users_before/len(df)*100:.1f}%)")
print(f"После уведомлений: {daily_users_after} ({daily_users_after/len(df)*100:.1f}%)")
```

Компания X тестирует продажу своего продукта через два разных сайта. На оба сайта приходит одинаковое количество пользователей. Причем выборки пользователей не пересекаются, то есть если пользователь увидел сайт 1, он уже не посетит сайт 2. Проверьте гипотезу, что средний чек по пользователям на сайте 1 отличается от среднего чека пользователей на сайте 2 (период сравнения — один месяц).

Опишите словами нулевую и альтернативную гипотезы. Проверьте гипотезы с помощью статистического критерия. Интерпретируйте получившиеся результаты. Уровень значимости = 5%.

Нулевая гипотеза: Средний чек пользователей на сайте 1 равен среднему чеку пользователей на сайте 2.

Альтернативная гипотеза: Средний чек пользователей на сайте 1 отличается от среднего чека пользователей на сайте 2.

```python
# Ваш код здесь

import pandas as pd
import numpy as np
from scipy import stats

print("=== АНАЛИЗ РАЗЛИЧИЙ СРЕДНЕГО ЧЕКА МЕЖДУ ДВУМЯ САЙТАМИ ===\n")

# Создаем примерные данные для анализа
np.random.seed(42)

# Данные для сайта 1
n_users = 100
site1_checks = np.random.normal(loc=2500, scale=800, size=n_users)
site1_checks = np.abs(site1_checks)  # чтобы не было отрицательных значений

# Данные для сайта 2 (немного другое распределение)
site2_checks = np.random.normal(loc=2700, scale=750, size=n_users)
site2_checks = np.abs(site2_checks)

# Создаем DataFrame
df = pd.DataFrame({
    'site1': site1_checks,
    'site2': site2_checks
})

print("Первые 10 строк данных:")
print(df.head(10))
print()

print("=== ОПИСАТЕЛЬНАЯ СТАТИСТИКА ===")
print(f"Сайт 1:")
print(f"  Средний чек: {df['site1'].mean():.2f} руб.")
print(f"  Стандартное отклонение: {df['site1'].std():.2f} руб.")
print(f"  Медиана: {df['site1'].median():.2f} руб.")
print(f"  Минимальный чек: {df['site1'].min():.2f} руб.")
print(f"  Максимальный чек: {df['site1'].max():.2f} руб.")
print(f"  Количество пользователей: {len(df['site1'])}")

print(f"\nСайт 2:")
print(f"  Средний чек: {df['site2'].mean():.2f} руб.")
print(f"  Стандартное отклонение: {df['site2'].std():.2f} руб.")
print(f"  Медиана: {df['site2'].median():.2f} руб.")
print(f"  Минимальный чек: {df['site2'].min():.2f} руб.")
print(f"  Максимальный чек: {df['site2'].max():.2f} руб.")
print(f"  Количество пользователей: {len(df['site2'])}")

print(f"\nРазница средних чеков (Сайт 2 - Сайт 1): {df['site2'].mean() - df['site1'].mean():.2f} руб.")

print(f"\n=== ФОРМУЛИРОВКА ГИПОТЕЗ ===")
print("Нулевая гипотеза (H₀): Средний чек пользователей на сайте 1 равен")
print("                       среднему чеку пользователей на сайте 2")
print("                       μ₁ = μ₂")
print("Альтернативная гипотеза (H₁): Средний чек пользователей на сайте 1")
print("                             отличается от среднего чека пользователей на сайте 2")
print("                             μ₁ ≠ μ₂")

print(f"\n=== ПРОВЕРКА ПРЕДПОЛОЖЕНИЙ ===")

# Проверяем нормальность распределения для обоих сайтов
shapiro_site1 = stats.shapiro(df['site1'])
shapiro_site2 = stats.shapiro(df['site2'])

print(f"Проверка нормальности распределения (тест Шапиро-Уилка):")
print(f"Сайт 1: p-value = {shapiro_site1.pvalue:.3f}")
print(f"Сайт 2: p-value = {shapiro_site2.pvalue:.3f}")

# Проверяем равенство дисперсий
levene_test = stats.levene(df['site1'], df['site2'])
print(f"\nТест Левене на равенство дисперсий:")
print(f"p-value = {levene_test.pvalue:.3f}")

print(f"\n=== ПРОВЕРКА ГИПОТЕЗ ===")

# Выбираем тип теста в зависимости от результатов проверки предположений
if shapiro_site1.pvalue > 0.05 and shapiro_site2.pvalue > 0.05:
    print("Распределения можно считать нормальными")
    
    if levene_test.pvalue > 0.05:
        print("Дисперсии можно считать равными -> используем t-тест Стьюдента")
        t_test = stats.ttest_ind(df['site1'], df['site2'], equal_var=True)
        test_used = "t-тест Стьюдента для независимых выборок"
    else:
        print("Дисперсии нельзя считать равными -> используем t-тест Уэлча")
        t_test = stats.ttest_ind(df['site1'], df['site2'], equal_var=False)
        test_used = "t-тест Уэлча для независимых выборок"
else:
    print("Распределения не нормальные -> используем U-тест Манна-Уитни")
    t_test = stats.mannwhitneyu(df['site1'], df['site2'], alternative='two-sided')
    test_used = "U-тест Манна-Уитни"

print(f"\nРезультаты статистического теста ({test_used}):")

if 'mannwhitneyu' in str(type(t_test)):
    print(f"U-статистика = {t_test.statistic:.4f}")
    print(f"p-value = {t_test.pvalue:.4f}")
else:
    print(f"t-статистика = {t_test.statistic:.4f}")
    print(f"p-value = {t_test.pvalue:.4f}")
    if hasattr(t_test, 'df'):
        print(f"Степени свободы = {t_test.df:.1f}")

print(f"\n=== ИНТЕРПРЕТАЦИЯ РЕЗУЛЬТАТОВ ===")
print("Уровень значимости: α = 0.05")

if t_test.pvalue < 0.05:
    print(f"p-value ({t_test.pvalue:.4f}) < 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМЫЕ РАЗЛИЧИЯ ОБНАРУЖЕНЫ")
    print("✓ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print("✓ ПРИНИМАЕМ альтернативную гипотезу H₁")
    print(f"\nВЫВОД: Средний чек пользователей статистически значимо")
    print(f"отличается между сайтом 1 и сайтом 2.")
    
    if df['site1'].mean() > df['site2'].mean():
        print(f"Сайт 1 показывает более высокий средний чек")
    else:
        print(f"Сайт 2 показывает более высокий средний чек")
        
else:
    print(f"p-value ({t_test.pvalue:.4f}) ≥ 0.05")
    print("✓ СТАТИСТИЧЕСКИ ЗНАЧИМЫХ РАЗЛИЧИЙ НЕ ОБНАРУЖЕНО")
    print("✓ НЕ ОТВЕРГАЕМ нулевую гипотезу H₀")
    print(f"\nВЫВОД: Нет статистически значимых различий в среднем чеке")
    print(f"между сайтом 1 и сайтом 2.")

print(f"\n=== ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ ===")

# Доверительные интервалы для средних
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
    return mean - h, mean + h

ci_site1 = confidence_interval(df['site1'])
ci_site2 = confidence_interval(df['site2'])

print(f"95% доверительный интервал для среднего чека:")
print(f"Сайт 1: [{ci_site1[0]:.2f}, {ci_site1[1]:.2f}] руб.")
print(f"Сайт 2: [{ci_site2[0]:.2f}, {ci_site2[1]:.2f}] руб.")

# Разница средних с доверительным интервалом
mean_diff = df['site2'].mean() - df['site1'].mean()
diff_std_err = np.sqrt(df['site1'].var()/len(df['site1']) + df['site2'].var()/len(df['site2']))
ci_diff = stats.t.interval(0.95, 
                          df=len(df['site1'])+len(df['site2'])-2, 
                          loc=mean_diff, 
                          scale=diff_std_err)

print(f"\nРазница средних чеков (Сайт 2 - Сайт 1): {mean_diff:.2f} руб.")
print(f"95% доверительный интервал для разницы: [{ci_diff[0]:.2f}, {ci_diff[1]:.2f}] руб.")

# Эффективность в процентах
if df['site1'].mean() > 0:
    percentage_diff = (mean_diff / df['site1'].mean()) * 100
    print(f"Относительная разница: {percentage_diff:.1f}%")

# Анализ пользователей по сегментам (условно)
def analyze_segments(data, segments):
    results = {}
    for name, threshold in segments.items():
        count = len([x for x in data if x >= threshold])
        percentage = (count / len(data)) * 100
        results[name] = (count, percentage)
    return results

segments = {
    'Низкий чек (<1500 руб.)': 1500,
    'Средний чек (1500-3500 руб.)': 1500,
    'Высокий чек (>3500 руб.)': 3500
}

print(f"\nРаспределение пользователей по сегментам:")
print(f"{'Сегмент':<30} {'Сайт 1':<15} {'Сайт 2':<15}")
print("-" * 60)

for segment, threshold in segments.items():
    if 'Низкий' in segment:
        count1 = len([x for x in df['site1'] if x < 1500])
        count2 = len([x for x in df['site2'] if x < 1500])
    elif 'Средний' in segment:
        count1 = len([x for x in df['site1'] if 1500 <= x <= 3500])
        count2 = len([x for x in df['site2'] if 1500 <= x <= 3500])
    else:  # Высокий
        count1 = len([x for x in df['site1'] if x > 3500])
        count2 = len([x for x in df['site2'] if x > 3500])
    
    perc1 = (count1 / len(df['site1'])) * 100
    perc2 = (count2 / len(df['site2'])) * 100
    
    print(f"{segment:<30} {count1} ({perc1:.1f}%)     {count2} ({perc2:.1f}%)")
```
